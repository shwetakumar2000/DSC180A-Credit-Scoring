```python
import pandas as pd
```

```python
raw_data_df = pd.read_csv('raw/cs-training.csv', index_col=0)
raw_data_df
```

```python
def check_na():
    num_columns = len(raw_data_df.columns)

    for i in range(num_columns):
        count = raw_data_df.iloc[:,i].isna().sum()
        print(raw_data_df.columns[i], count, count / len(raw_data_df))
        
check_na()
```

```python
# Extract Debt amount from DebtRatio and MontlyIncome
raw_data_df = raw_data_df[raw_data_df["MonthlyIncome"].notna()]
raw_data_df["Debt"] = raw_data_df["MonthlyIncome"] * raw_data_df["DebtRatio"]
raw_data_df
```
